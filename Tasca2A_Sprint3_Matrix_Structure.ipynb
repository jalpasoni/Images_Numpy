Create a two-dimensional array with a shape of 5 x 5. Extract the maximum value of the array, and the maximum values of each of its axes.

###  Exercise 5 

Use Indexing to extract the values of a column and a row from the array. And add up their values.


```python
arr = np.arange(16).reshape((4, 4))
arr
```

```python
# for columns
nc = 2
col_n = arr[:, nc]
print('Column ' + str(nc) + ':', col_n)
#sum_col = col_n.sum()
print('Sum of elements = ', col_n.sum())

print('')

# for rows
nr = 0
row_n = arr[nr, :]
print('Row ' + str(nr) + ':', row_n)
print('Sum of elements = ', row_n.sum())
```

**_Let's add some shades to the image_**

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(12, 8))
ax1.axis("off")
ax2.axis("off")
ax3.axis("off")
ax4.axis("off")
ax1.imshow(my_img * (0.002))
ax2.imshow(my_img * (0.004))
ax3.imshow(my_img * (0.008))
ax4.imshow(my_img * (0.020))

fig.savefig('../images/shaded.png', bbox_inches='tight')
```

### Exercises 8

Show what happens when we remove the Green G or Blue B channel. You should use indexing to select the channel you want to undo.

Use the method, mpimg.imsave () from the imported library, to save the modified images and you will need to upload them to your repository on github.

```python
remove_green = my_img.copy() 
remove_green[:,:,1] = 0
remove_blue = my_img.copy() 
remove_blue[:,:,2] = 0
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))
ax1.axis("off")
ax2.axis("off")
ax1.imshow(remove_green)
ax2.imshow(remove_blue)
ax1.set_title('Red + Blue', fontsize=18, color='purple')
ax2.set_title('Red + Green', fontsize=18, color='orange')

fig.savefig('../images/NoGreen_NoBlue.png', bbox_inches='tight')
```
